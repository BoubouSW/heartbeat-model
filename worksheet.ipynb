### Solution du système Van der Pol dans le cas où on ne rajoute pas de perturbation x(t)²-1

```python
lamb, mu = 1, 1

tt = np.linspace(0,10,200)

x = lambda t: lamb*np.cos(t) + mu*np.sin(t)
y = lambda t: -lamb*np.sin(t) + mu*np.cos(t)

fig = plt.figure(figsize=(6,6))

ax = fig.add_subplot(1,1,1)

ax.plot(x(tt),y(tt));
ax.set_xlabel("x(t)")
ax.set_ylabel("y(t)")
```

On trouve une solution cyclique : c'est pour cela qu'on rajoute une perturbation pour ne pas être dans ce cas
